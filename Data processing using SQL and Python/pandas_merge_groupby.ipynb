### Задача 4. Работа с ФИО

Создай три дополнительные колонки в датафрейме, которые будут содержать фамилию, имя и отчество.

Если что-то начинается не с заглавной буквы, переведи первую букву в верхний регистр.

**Подсказка:**

 1. Изучи метод **strip** и воспользуйся им для удаления лишних пробелов в начале и конце строк. [Ссылка на описание](https://ru.hexlet.io/qna/python/questions/chto-delaet-funktsiya-strip-v-python).
 2. Убедись, что никакие символы, кроме пробелов, не используются для разделения фамилии, имени и отчества. Если используются, то замени их на пробелы.
 3. Воспользуйся методом **split** для разбиения по разделителю и запиши получившиеся части соответственно в три колонки.
 4. Поменяй регистр первых букв. Для этого можно воспользоваться методом **upper**.

```python
### Решение тут
def complex_capitalize(text: str) -> str:
  if text == '':
    return text
  if '-' in text:
    splitted = text.split('-')
    complex_part = []
    for part in splitted:
      if part:
        complex_part.append(part[0].upper() + part[1:].lower())
    return('-'.join(complex_part))
  else:
    return (text[0].upper() + text[1:].lower())


def correct_text(text: str) -> tuple:
  if pd.isna(text):
    return (None, None, None)
  text = str(text).strip()
  cleaned_text = re.sub(r'[^a-zA-Zа-яА-ЯёЁ\s\-]', ' ', text)
  splitted_text = cleaned_text.split()
  if len(splitted_text) == 0:
    return (None, None, None)
  elif len(splitted_text) == 1:
    surname = complex_capitalize(splitted_text[0])
    return (surname, None, None)
  elif len(splitted_text) == 2:
    surname = complex_capitalize(splitted_text[0])
    name = complex_capitalize(splitted_text[1])
    return (surname, name, None)
  else:
    surname = complex_capitalize(splitted_text[0])
    name = complex_capitalize(splitted_text[1])
    patronymic = ' '.join(complex_capitalize(part) for part in splitted_text[2:])
    return (surname, name, patronymic)


cleaning[['surname','first_name','patronymic']] = cleaning['name'].apply(lambda x: pd.Series(correct_text(x)))
cleaning
```

### Задание 1

Правда ли, что:

 - Поле *id_purchase* является *primary key* в таблице **orders**? (Проверь уникальность значений, а также наличие нуллов в колонке *id_purchase* в таблице **orders**).
 - Поле *id_client* является *primary key* в таблице **clients**?  (Проверь уникальность значений, а также наличие нуллов в колонке *id_client* в таблице **clients** ).
 - В таблице **orders** в полях *id_client*, *sum_pay*, *id_trading_point*, *date_pay* нет пропусков (нуллов)?
 - В таблице **clients** в полях *gender* и *age* нет пропусков (нуллов)?

Результаты выведи с помощью интерпретируемых (легко читаемых) принтов.

```python
# your code here
def check_primary_key(df:pd.DataFrame, df_name:str, field:str) -> None:
  is_unique = df[field].is_unique
  has_null = df[field].isnull().any()
  count_null = df[field].isnull().sum()
  count_duplicates = df[field].duplicated().sum()
  is_primary_key = is_unique and not has_null
  print(f'{field} {"является" if is_primary_key else "не является"} primary key в {df_name}')
  if not is_primary_key:
    print(f'Всего значений в {df_name} -> {len(df)}\nКоличество NULL в {field} -> {count_null}\nПроцент NULL значений в поле {field} -> {count_null/len(df) * 100:.3f}%\nКоличество дубликатов в {field} -> {count_duplicates}\nПроцент дубликатов в поле {field} -> {count_duplicates/len(df) * 100:.3f}%\n')

def check_nulls(df:pd.DataFrame, df_name:str, field:str) -> None:
  has_null = df[field].isnull().any()
  count_null = df[field].isnull().sum()
  print(f'{field} {"имеет" if has_null else "не имеет"} NULL в {df_name}')
  if has_null:
    print(f'Всего значений в {df_name} -> {len(df)}\nКоличество NULL в {field} -> {count_null}\nПроцент NULL значений в поле {field} -> {count_null/len(df) * 100:.3f}%\n')


check_primary_key(df=df_orders, df_name='df_orders', field='id_purchase')
check_primary_key(df=df_clients, df_name='df_clients', field='id_client')

check_nulls(df=df_orders, df_name='df_orders', field='id_client')
check_nulls(df=df_orders, df_name='df_orders', field='sum_pay')
check_nulls(df=df_orders, df_name='df_orders', field='id_trading_point')
check_nulls(df=df_orders, df_name='df_orders', field='date_pay')

check_nulls(df=df_clients, df_name='df_clients', field='gender')
check_nulls(df=df_clients, df_name='df_clients', field='age')
```

```python
# your code here
def missing_client(df1:pd.DataFrame, df2:pd.DataFrame, field:str) -> list:
  return df1[~df1[field].isin(df2[field])][field].unique().tolist()


client_only_in_clients = missing_client(df1=df_clients, df2=df_orders, field='id_client')
client_only_in_orders = missing_client(df1=df_orders, df2=df_clients, field='id_client')

print(f'Количество клиентов, которые есть только в таблице clients -> {len(client_only_in_clients)}')
print(f'Количество клиентов, которые есть только в таблице orders -> {len(client_only_in_orders)}')
```

### Задание 4

Правда ли, что средний чек по женщинам в возрасте от 20 до 30 лет выше среднего чека по мужчинам в этой же возрастной категории?

В ответе выведи, на сколько рублей (0 знаков после запятой) и на сколько процентов (1 знак после запятой) женский чек выше мужского.

```python
# your code here
merged_pay_by_gender_age = df_clients.merge(df_orders[['id_client','sum_pay']], how='inner', on='id_client')
mean_pay_by_gender_age = (merged_pay_by_gender_age[merged_pay_by_gender_age['age'].between(20,30)]
                          .groupby('gender').agg(mean_by_gender=('sum_pay','mean'))
                          )
female_avg = mean_pay_by_gender_age.loc['F']
male_avg = mean_pay_by_gender_age.loc['M']
diff = round(female_avg - male_avg).iloc[0]
percent = ((female_avg - male_avg)/male_avg * 100).iloc[0]
print(f'{"Гипотеза верна" if diff > 0 else "Гипотеза ложна"}')
print(f'Разница в чеках -> {diff:.0f} у.е. или {percent:.1f}%')
```

Предположим, что стоимость обработки одной операции в торговой точке (с учётом костов на аренду, продавца и так далее) составляет $250$ рублей.

Для каждой торговой точки рассчитай:
 - количество операций;
 - суммарную выручку;
 - чистую валовую прибыль;
 - валовый средний чек (из расчёта на одну операцию);
 - чистый средний чек (то есть валовый средний чек из расчёта на одну операцию за вычетом костов);
 - уникальное количество покупающих клиентов;
 - валовый средний чек из расчёта на одного клиента.

```python
df_orders.head()
```

```python
cost = 250
trading_point_stats = (
    df_orders.groupby('id_trading_point')
    .agg(operations_count=('id_purchase','count'),
         revenue=('sum_pay','sum'),
         unique_clients=('id_client','nunique')
         )
)
trading_point_stats['gross_profit'] = trading_point_stats['revenue'] - cost * trading_point_stats['operations_count']
trading_point_stats['avg_gross_check'] = trading_point_stats['revenue'] / trading_point_stats['operations_count']
trading_point_stats['avg_net_check'] = trading_point_stats['avg_gross_check'] - cost
trading_point_stats['avg_check_per_client'] = trading_point_stats['revenue'] / trading_point_stats['unique_clients']
```
